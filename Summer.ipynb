# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
dataFile = pd.read_csv("summer.csv")
len(dataFile['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
sports = dataFile['Sport'].unique()
sport_medal_count = []
for sport in sports:
    x = dataFile[dataFile['Sport'] == sport]
    sport_medal_count.append([sport,len(x[x['Medal'] == 'Gold'])])
pd.DataFrame(sport_medal_count, columns=['Sport','MedalCount']).sort_values(by='MedalCount',ascending=False).head().plot(x='Sport', y='MedalCount', kind='bar')    
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
sports = dataFile['Sport'].unique()
sport_medal_count = []
for sport in sports:
    x = dataFile[dataFile['Sport'] == sport]
    sport_medal_count.append([sport,len(x)])
pd.DataFrame(sport_medal_count, columns=['Sport','MedalCount']).sort_values(by='MedalCount',ascending=False).head().plot(x='Sport', y='MedalCount', kind='bar')
```

### 4. Which player has won most number of medals? (Top 5)

```python
athletes = dataFile['Athlete'].unique()
medal_count = []
for athlete in athletes:
    x = dataFile[dataFile['Athlete'] == athlete]
    medal_count.append([athlete,len(x)])
pd.DataFrame(medal_count, columns=['Athlete','MedalCount']).sort_values(by='MedalCount',ascending=False).head().plot(x='Athlete', y='MedalCount', kind='bar')    
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
indianMedals = dataFile[dataFile['Country'] == 'IND']
indianMedalsYears = indianMedals[indianMedals['Medal'] == 'Gold']
years = []
for year in indianMedalsYears['Year']:
    years.append(year)
print(min(years))
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
events = dataFile['Event'].unique()
eventPlayers = []
for event in events:
        x = dataFile[dataFile['Event'] == event]
        eventPlayers.append([event,len(x)])
pd.DataFrame(eventPlayers, columns=['Event','PlayerCount']).sort_values(by='PlayerCount',ascending=False).head().plot(x='Event', y='PlayerCount', kind='bar') 
```
